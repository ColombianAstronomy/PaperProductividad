```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
names = ['Colombia', 'Uniandes', 'UNal', 'UdeA', 'UIS']

data = {}
for i, name in enumerate(names):
    data[name] = pd.read_csv('../data/historia_{}_ADS.csv'.format(name))
```

```python
widths = [6,3,3,3,3]
alphas = [1, 1, 0.3, 1, 0.3]
colors = ['black', 'black', 'black', 'black', 'black']
linestyles = ['solid', 'solid', 'solid', 'dashed', 'dashed']
for i, name in enumerate(names):
    plt.plot(data[name]['Year'], np.cumsum(data[name][' Article Count']), 
             linewidth=widths[i], 
             label=name, 
             color=colors[i],
             alpha=alphas[i],
             linestyle=linestyles[i])
plt.yscale('log')
plt.legend()
plt.grid()
plt.xlabel('Ao')
plt.ylabel('# Publicaciones acumuladas')
```
