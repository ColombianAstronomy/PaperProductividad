```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
names = ['Colombia', 'Uniandes', 'UNal', 'UdeA', 'UIS']

data = {}
for i, name in enumerate(names):
    data[name] = pd.read_csv('../data/historia_{}_ADS.csv'.format(name))
```

```python
widths = [8,6,5,3,2]
alphas = [1,0.9,0.7, 0.6, 0.5]
for i, name in enumerate(names):
    plt.plot(data[name]['Year'], np.cumsum(data[name][' Article Count']), linewidth=widths[i], 
             label=name, alpha=alphas[i])
plt.yscale('log')
plt.legend()
plt.grid()
plt.xlabel('Año')
plt.ylabel('# Publicaciones acumuladas')
```
